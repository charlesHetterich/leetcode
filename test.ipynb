```python
class Solution:
    def nDigits(self, n: int) -> int:
        L = 0
        while n >= 1:
            n /= 10
            L += 1
        return L

    def step(self, c: int, l: int, n: int, L: int) -> tuple[int, int]:
        if l < L and c * 10 <= n:               # add 0 digit
            c *= 10
            l += 1
        else:
            if c % 10 < 9:                      # add 1 to end of c
                c += 1
                if c % 10 == 9 and c <= n:
                    return c, l
            
            did_cut = c % 10 == 9 or c > n
            while c % 10 == 9 or c > n:         # remove digits and add 1 to end of c
                c //= 10
                l -= 1
            if did_cut:
                c += 1
        return c, l
    
    def skip(self, c: int, l: int, n: int, L: int) -> tuple[int, int]:
        return c,c

    def findKthNumber(self, n: int, k: int) -> int:
        L = self.nDigits(n)
        
        # current digits and value
        c, l, kk = 1, 1, 1

        outs = [(c, kk)]
        for _ in range(k - 1):
            c, l = self.step(c, l, n, L)
            kk += 1

            outs.append((c, kk))
        return outs
```
